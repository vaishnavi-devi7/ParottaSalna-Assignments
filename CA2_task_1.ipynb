```python
# 1. Create dictionary student and print it

student = {
    "name": "Alice",
    "age": 21,
    "major": "Computer Science"
}
print("Q1 student:", student)

# 2. Print values of "name" and "major"

print("Q2 name:", student["name"])
print("Q2 major:", student["major"])

# 3. Add "gpa": 3.8 and update "age" to 22

student["gpa"] = 3.8
student["age"] = 22
print("Q3 updated student:", student)

# 4. Remove "major" using del and print

del student["major"]
print("Q4 after removing 'major':", student)

# 5. Check if "age" key exists

print("Q5 'age' in student?:", "age" in student)

# 6. Create prices dictionary and print each key-value pair

prices = {
    "apple": 0.5,
    "banana": 0.3,
    "orange": 0.7
}
print("Q6 prices key-value pairs:")
for item, price in prices.items():
    print(item, "=", price)

# 7. Use len() to find number of key-value pairs in prices

print("Q7 len(prices):", len(prices))

# 8. Use get() for "gpa" and non-existing "graduation_year" with default 2025

print("Q8 gpa:", student.get("gpa"))
print("Q8 graduation_year (with default):", student.get("graduation_year", 2025))

# 9. Create extra_info and merge into student with update()

extra_info = {
    "graduation_year": 2025,
    "hometown": "Springfield"
}
student.update(extra_info)
print("Q9 merged student:", student)

# 10. Create dictionary squares using comprehension

squares = {n: n**2 for n in range(1, 6)}
print("Q10 squares:", squares)

# 11. Print keys and values of prices as separate lists

price_keys = list(prices.keys())
price_values = list(prices.values())
print("Q11 price keys:", price_keys)
print("Q11 price values:", price_values)

# 12. Create nested dict school and print age of student2

school = {
    "student1": {"name": "Alice", "age": 21},
    "student2": {"name": "Bob", "age": 22}
}
print("Q12 age of student2:", school["student2"]["age"])

# 13. Use setdefault() to add "advisor" if not exists

student.setdefault("advisor", "Dr. Smith")
print("Q13 student with advisor:", student)

# 14. Use pop() to remove "hometown" and store its value

hometown_value = student.pop("hometown")
print("Q14 popped hometown:", hometown_value)
print("Q14 student after popping hometown:", student)

# 15. Use clear() on prices and print

prices.clear()
print("Q15 cleared prices:", prices)

# 16. Make a copy of student and modify copy

student_copy = student.copy()
student_copy["name"] = "Charlie"
print("Q16 original student:", student)
print("Q16 modified copy:", student_copy)

# 17. Use zip() to create dict from lists

keys = ["name", "age", "major"]
values = ["Eve", 20, "Mathematics"]
new_student = dict(zip(keys, values))
print("Q17 new_student from zip:", new_student)

# 18. Use items() to iterate over student and print

print("Q18 student items:")
for key, value in student.items():
    print(key, "=", value)

# 19. fruit_count: count occurrences of each fruit

fruits = ["apple", "banana", "apple", "orange", "banana", "banana"]
fruit_count = {}
for f in fruits:
    fruit_count[f] = fruit_count.get(f, 0) + 1
print("Q19 fruit_count:", fruit_count)

# 20. Use collections.defaultdict for word_count

from collections import defaultdict

words = ["hello", "world", "hello", "python"]
word_count = defaultdict(int)
for w in words:
    word_count[w] += 1
print("Q20 word_count (defaultdict):", dict(word_count))
```
