```python
#Create list and print third item

items = ["Notebook", "Pencil", "Eraser", "Ruler", "Marker"]
print("Q1 third item:", items[2])
```

```python
#Add "Glue Stick" to end and print

items.append("Glue Stick")
print("Q2 list:", items)
```

```python
#Insert "Highlighter" between 2nd and 3rd and print

items.insert(2, "Highlighter")
print("Q3 list:", items)
```

```python
#Remove "Ruler" and print

items.remove("Ruler")
print("Q4 list:", items)
```

```python
#Print sublist with first three items

sub_list = items[:3]
print("Q5 sublist:", sub_list)
```

```python
#Convert all item names to uppercase using list comprehension

upper_items = [item.upper() for item in items]
print("Q6 uppercase list:", upper_items)
```

```python
#Check if "Marker" is still in the list

if "Marker" in items:
    print("Q7: Marker is in the list.")
else:
    print("Q7: Marker is NOT in the list.")
```

```python
#Print number of delivery items

print("Q8 count:", len(items))
```

```python
#Sort list alphabetically and print

sorted_items = sorted(items)
print("Q9 sorted:", sorted_items)
```

```python
#Reverse the order and print

reversed_items = list(reversed(items))
print("Q10 reversed:", reversed_items)
```

```python
#List of [item, time] and print first item and time

deliveries_with_time = [
    ["Notebook", "10:00 AM"],
    ["Pencil", "10:15 AM"],
    ["Highlighter", "10:30 AM"],
    ["Eraser", "10:45 AM"],
    ["Marker", "11:00 AM"],
]
print("Q11 first item and time:", deliveries_with_time[0])
```

```python
#Count how many times "Ruler" appears

print("Q12 Ruler count:", items.count("Ruler"))
```

```python
#Find index of "Pencil"

print("Q13 index of Pencil:", items.index("Pencil"))
```

```python
#Extend list with another list and print

new_items = ["Stapler", "Tape", "Clip"]
items.extend(new_items)
print("Q14 extended list:", items)
```

```python
#Clear list and print

items.clear()
print("Q15 cleared list:", items)
```

```python
#List with "Notebook" repeated three times

notebooks = ["Notebook"] * 3
print("Q16 repeated list:", notebooks)
```

```python
#Nested list comprehension: [item, length]

items2 = ["Notebook", "Pencil", "Eraser", "Ruler", "Marker", "Glue Stick", "Highlighter"]
item_lengths = [[item, len(item)] for item in items2]
print("Q17 item and length list:", item_lengths)
```

```python
#Filter items containing letter "e"

filtered_items = [item for item in items2 if "e" in item or "E" in item]
print("Q18 filtered (contains 'e'):", filtered_items)
```

```python
#Remove duplicates from list

items_with_dupes = ["Notebook", "Pencil", "Eraser", "Notebook", "Ruler", "Pencil"]
unique_items = list(dict.fromkeys(items_with_dupes))
print("Q19 unique items:", unique_items)
```
