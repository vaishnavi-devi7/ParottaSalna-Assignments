```python
# Create a set named rose_garden containing different types of roses: "red rose", "white rose", "yellow rose". Print the same.

rose_garden = {"red rose", "white rose", "yellow rose"}
print("1.rose_garden:", rose_garden)

# Add "pink rose" to the rose_garden set. Print the set to confirm the addition.

rose_garden.add("pink rose")
print("2.rose_garden after adding pink rose:", rose_garden)

# Remove "yellow rose" from the rose_garden set using the remove() method. Print the set to verify the removal.

rose_garden.remove("yellow rose")
print("3.rose_garden after removing yellow rose:", rose_garden)

# Create another set botanical_garden with elements "sunflower", "tulip", and "red rose". Find the union of rose_garden and botanical_garden and print the result.

botanical_garden = {"sunflower", "tulip", "red rose"}
union_set = rose_garden.union(botanical_garden)
print("4.union of rose_garden and botanical_garden:", union_set)

# Find the intersection of rose_garden and botanical_garden and print the common elements.

intersection_set = rose_garden.intersection(botanical_garden)
print("5.intersection of rose_garden and botanical_garden:", intersection_set)

# Find the difference between rose_garden and botanical_garden and print the elements that are only in rose_garden.

difference_set = rose_garden.difference(botanical_garden)
print("6.elements only in rose_garden:", difference_set)

# Find the symmetric difference between rose_garden and botanical_garden and print the elements unique to each set.

sym_diff_set = rose_garden.symmetric_difference(botanical_garden)
print("7.symmetric difference:", sym_diff_set)

# Create a set small_garden containing "red rose", "white rose". Check if small_garden is a subset of rose_garden and print the result.

small_garden = {"red rose", "white rose"}
print("8.small_garden is subset of rose_garden?:", small_garden.issubset(rose_garden))

# Check if rose_garden is a superset of small_garden and print the result.

print("9.rose_garden is superset of small_garden?:", rose_garden.issuperset(small_garden))

# Use the len() function to find the number of elements in the rose_garden set. Print the result.

print("10.number of elements in rose_garden:", len(rose_garden))

# Use the discard() method to remove "pink rose" from the rose_garden set. Try to discard a non-existent element "blue rose" and observe what happens.

rose_garden.discard("pink rose")
rose_garden.discard("blue rose")
print("11.rose_garden after discarding pink rose and blue rose:", rose_garden)

# Use the clear() method to remove all elements from the rose_garden set. Print the set to confirm it’s empty.

rose_garden.clear()
print("12.rose_garden after clear():", rose_garden)

# Make a copy of the botanical_garden set using the copy() method. Add "lily" to the copy and print both sets to see the differences.

botanical_copy = botanical_garden.copy()
botanical_copy.add("lily")
print("13.original botanical_garden:", botanical_garden)
print("13.copied botanical_garden with lily:", botanical_copy)

# Create a frozen set immutable_garden with elements "orchid", "daisy", "red rose". Try to add or remove an element and observe what happens.

immutable_garden = frozenset({"orchid", "daisy", "red rose"})
print("14.immutable_garden (frozenset, cannot be changed):", immutable_garden)



# Iterate over the botanical_garden set and print each element.

print("15.elements of botanical_garden:")
for flower in botanical_garden:
    print(flower)

# Use set comprehension to create a set even_numbers containing even numbers from 1 to 10.

even_numbers = {n for n in range(1, 11) if n % 2 == 0}
print("16.even_numbers from 1 to 10:", even_numbers)

# Given a list of flowers ["rose", "tulip", "rose", "daisy", "tulip"], use a set to remove duplicates and print the unique flowers.

flowers = ["rose", "tulip", "rose", "daisy", "tulip"]
unique_flowers = set(flowers)
print("18.unique flowers:", unique_flowers)

# Check if "sunflower" is in the botanical_garden set and print the result.

print('"19.sunflower" in botanical_garden?:', "sunflower" in botanical_garden)

# Use the intersection_update() method to update the botanical_garden set with only the elements found in rose_garden. Print the updated set.

rose_garden = {"red rose", "white rose", "yellow rose", "pink rose"}
botanical_garden.intersection_update(rose_garden)
print("20.botanical_garden after intersection_update with rose_garden:", botanical_garden)

# Use the difference_update() method to remove all elements in small_garden from botanical_garden. Print the updated set.

botanical_garden.difference_update(small_garden)
print("21.botanical_garden after difference_update with small_garden:", botanical_garden)
```
