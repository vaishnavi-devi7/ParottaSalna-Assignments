```python
# How do you use an f-string to print name, age, and city in the format "Name: …, Age: …, City: …"?

name, age, city = "vaishnavi", 20, "Chennai"
print(f"Name: {name}, Age: {age}, City: {city}")
```

```python
# Print using the .format() method?

name = "vaishnavi "
age = 20
city = "Chennai"
print("Name: {}, Age: {}, City: {}".format(name, age, city))
```

```python
# Print pi rounded to 2 decimal places?

pi = 3.14159
print(f"The value of pi is approximately {pi:.2f}")
```

```python
# Print "left" left-aligned and "right" right-aligned within 10 characters?

print("{:<10}".format("left") + "{:>10}".format("right"))
```
