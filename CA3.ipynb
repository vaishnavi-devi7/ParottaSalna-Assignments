```python
import sqlite3

conn = sqlite3.connect("leaderboard.db")
cursor = conn.cursor()

def show_leaderboard(sort_by="difficulty", order="ASC"):

    query = f"""
        SELECT player_name, difficulty, attempts
        FROM leaderboard
        ORDER BY {sort_by} {order}
    """
    cursor.execute(query)
    rows = cursor.fetchall()
    print(f"\n--- Leaderboard sorted by {sort_by} ({order}) ---")
    for name, diff, att in rows:
        print(f"Name: {name}, Difficulty: {diff}, Attempts: {att}")

def leaderboard_menu():
    while True:
        print("\n1. Show leaderboard")
        print("2. Show leaderboard sorted by difficulty (ASC)")
        print("3. Show leaderboard sorted by difficulty (DESC)")
        print("4. Show leaderboard sorted by attempts (ASC)")
        print("5. Show leaderboard sorted by attempts (DESC)")
        print("6. Exit leaderboard menu")

        choice = input("Enter choice: ")

        if choice == "1":
            show_leaderboard()
        elif choice == "2":
            show_leaderboard(sort_by="difficulty", order="ASC")
        elif choice == "3":
            show_leaderboard(sort_by="difficulty", order="DESC")
        elif choice == "4":
            show_leaderboard(sort_by="attempts", order="ASC")
        elif choice == "5":
            show_leaderboard(sort_by="attempts", order="DESC")
        elif choice == "6":
            break
        else:
            print("Invalid choice. Try again.")
```
